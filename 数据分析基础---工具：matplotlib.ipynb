### 一、基本用法
- 首先分解要绘制图像的相关组成元素，然后逐步完成绘图即可。


```python
# 可以通过“方法+？”查看参数及相关解释说明
plt.figure?
```

```python
# 创建数据
x = np.arange(10)
y1 = 2*x + 1
y2 = x**2
# 定义图像窗口及绘图
plt.figure()#可以设置像素和图像大小
plt.plot(x,y2)
plt.plot(x,y1)
#plt.show()
#定义坐标轴范围及名称
# plt.xlim((0, 10))
# plt.ylim((0, 100))
# plt.xlabel('x')
# plt.ylabel('y')
#定义坐标轴刻度及名称
plt.xticks(np.linspace(0,10 , 5))
plt.yticks(np.linspace(0,100 , 5),[r'a', r'b', r'c', r'd', r'e'])
#plt.show()
#设置图像边框
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#plt.show()
#调整刻度及边框位置
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 50))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',5))
plt.show()
# 保存图片
plt.savefig("./hhh.svg")# 输入路径 ，默认格式是png，可以设置为矢量图(.svg)
```

```python
"""
图例与注释
"""
# 创建数据
x = np.arange(10)
y1 = 2*x + 1
y2 = x**2
# 定义图像窗口及绘图
plt.figure()
# 首先图中的线条在绘图时需要命名
l1 = plt.plot(x, y1, label='linear line')
l2 = plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--', label='square line')
#对图例进行一些个性化的设置
plt.legend(loc='upper right')#设置位置
plt.show()
```

### 二、常见图像的绘制


- 只需传入数据即可进行绘图，其他相关设置遵循基本用法，具体参数可以+？进行查看

```python
X = np.random.normal(0, 1, 10000) # 每一个点的X值
Y = np.random.normal(0, 1, 10000) # 每一个点的Y值
plt.scatter(X, Y, s=2, alpha=1)
plt.xlim(-1.5, 1.5)
plt.xticks(())  # ignore xticks
plt.ylim(-1.5, 1.5)
plt.yticks(())  # ignore yticks
plt.show()
```
